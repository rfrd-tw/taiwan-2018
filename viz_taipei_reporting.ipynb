```python
pat = re.compile('(\d{1,2}\/\d{1,2}) (\d{2}:\d{2}:\d{2})')


df_timestamps["timestamp"] = None

for i in range(min(len(df_timestamps), float('inf'))):
    rawtime = df_timestamps.loc[i,"完成登錄時間"]
#     print(df_timestamps.loc[i,"完成登錄時間"])
    results = re.findall(pat, rawtime)
#     print(results)
    df_timestamps.loc[i, "timestamp"] = pd.to_datetime("2018/"+results[0][0] + " " + results[0][1], format="%Y/%m/%d %H:%M:%S")
    
df_timestamps
    
```

```python
df_timestamps = df_timestamps.query("縣市 == '臺北市'").sort_values(by='timestamp')
df_timestamps
```

```python
df_results = pd.read_csv("rawdata/cec_results_taipei.csv", skiprows=1, thousands=',')
df_results
```

```python
df_new = df_timestamps.merge( df_results[["村里別", "投開票所別", "吳蕚洋", "丁守中", "姚文智", "柯文哲", "李錫錕", "有效票數"]], left_on="投開票所", right_on="投開票所別")
df_new
```

```python
df_cumsum = df_new[["吳蕚洋", "丁守中", "姚文智", "柯文哲", "李錫錕", "有效票數"]].cumsum()
# df_cumsum = pd.to_numeric( df_cumsum[:,[:]] )
df_cumsum
```

```python
df_cumsum = pd.concat([df_new[["timestamp"]], df_cumsum], axis=1)
df_cumsum
```

```python
vote_data = [ [], [], [], [], [] ]

for i in range(5):
    for j, tstamp in enumerate(df_cumsum["timestamp"]):
        vote_data[i].append( {"date": tstamp.strftime("%Y%m%d-%H%M%S"), "data": 1.0*df_cumsum.iloc[j,i+1]} )

vote_data
```

```python
with open('data/{}_cec_results_by_time.json'.format("taipei"), 'w') as outfile:
    json.dump(vote_data, outfile, indent = 4)
```
